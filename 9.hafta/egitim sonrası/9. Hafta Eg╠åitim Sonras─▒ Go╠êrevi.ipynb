# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_acik = False
        self.ses = 10
        self.kanallar = ["TRT", "ATV", "Kanal D", "Star", "NTV"]
        self.aktif_kanal = self.kanallar[0]

    def tv_ac(self):
        if not self.tv_acik:
            print("TV açılıyor...")
            time.sleep(1)
            self.tv_acik = True
            print("TV açık.")
        else:
            print("TV zaten açık.")

    def tv_kapat(self):
        if self.tv_acik:
            print("TV kapanıyor...")
            time.sleep(1)
            self.tv_acik = False
            print("TV kapalı.")
        else:
            print("TV zaten kapalı.")

    def ses_ayarlari(self):
        while True:
            secim = input("Sesi artır: '>' | Sesi azalt: '<' | Çıkış: 'x'\nSeçim: ")
            if secim == '>':
                if self.ses < 100:
                    self.ses += 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten maksimumda.")
            elif secim == '<':
                if self.ses > 0:
                    self.ses -= 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten minimumda.")
            elif secim == 'x':
                print("Ses ayarından çıkılıyor.")
                break
            else:
                print("Geçersiz seçim.")

    def kanal_ekle(self, yeni_kanal):
        if yeni_kanal not in self.kanallar:
            self.kanallar.append(yeni_kanal)
            print(f"{yeni_kanal} kanalı eklendi.")
        else:
            print("Bu kanal zaten mevcut.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgele_kanal(self):
        self.aktif_kanal = random.choice(self.kanallar)
        print(f"Rastgele geçilen kanal: {self.aktif_kanal}")

    def tv_bilgileri(self):
        print(f"TV Durumu: {'Açık' if self.tv_acik else 'Kapalı'}")
        print(f"Ses Düzeyi: {self.ses}")
        print(f"Kanallar: {', '.join(self.kanallar)}")
        print(f"Aktif Kanal: {self.aktif_kanal}")


def menu():
    kumanda = Kumanda()

    while True:
        print("\n--- KUMANDA MENÜSÜ ---")
        print("1. TV Aç")
        print("2. TV Kapat")
        print("3. Ses Ayarları")
        print("4. Kanal Ekle")
        print("5. Kanal Sayısını Öğren")
        print("6. Rastgele Kanala Geç")
        print("7. TV Bilgileri")
        print("q. Çıkış")

        secim = input("İşlem seçiniz: ")

        if secim == "1":
            kumanda.tv_ac()
        elif secim == "2":
            kumanda.tv_kapat()
        elif secim == "3":
            kumanda.ses_ayarlari()
        elif secim == "4":
            kanal = input("Eklemek istediğiniz kanal adı: ")
            kumanda.kanal_ekle(kanal)
        elif secim == "5":
            kumanda.kanal_sayisi()
        elif secim == "6":
            kumanda.rastgele_kanal()
        elif secim == "7":
            kumanda.tv_bilgileri()
        elif secim.lower() == "q":
            print("Kumanda kapatılıyor...")
            break
        else:
            print("Geçersiz seçim, tekrar deneyin.")


if __name__ == "__main__":
    menu()
```
