# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_sinifi(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_kaydet(dosya_adi, veri_listesi):
    with open(dosya_adi, "w", encoding="utf-8") as dosya:
        dosya.write("İsim Soyisim, Yaş, Boy (m), Kilo (kg), VKİ, Sınıf\n")
        for veri in veri_listesi:
            dosya.write(f"{veri['isim']} {veri['soyisim']}, {veri['yas']}, {veri['boy']:.2f}, {veri['kilo']}, {veri['vki']:.2f}, {veri['sinif']}\n")

def main():
    veri_listesi = []
    adet = int(input("Kaç kişi için giriş yapacaksınız? "))

    for i in range(adet):
        print(f"\n{i + 1}. kişi bilgileri:")
        isim = input("İsim: ").strip().capitalize()
        soyisim = input("Soyisim: ").strip().capitalize()
        yas = int(input("Yaş: "))
        boy = float(input("Boy (metre cinsinden, örn: 1.75): "))
        kilo = float(input("Kilo (kg cinsinden): "))

        vki = vki_hesapla(kilo, boy)
        sinif = vki_sinifi(vki)

        print(f"{isim} {soyisim} için VKİ: {vki:.2f} - {sinif}")

        veri_listesi.append({
            "isim": isim,
            "soyisim": soyisim,
            "yas": yas,
            "boy": boy,
            "kilo": kilo,
            "vki": vki,
            "sinif": sinif
        })

    dosya_adi = "vki_kayitlari.txt"
    veri_kaydet(dosya_adi, veri_listesi)
    print(f"\nTüm veriler '{dosya_adi}' adlı dosyaya kaydedildi.")

if __name__ == "__main__":
    main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python


menu = {
    "Çilek": 15,
    "Muz": 10,
    "Yaban Mersini": 18,
    "Ispanak": 8,
    "Badem Sütü": 12,
    "Yulaf": 7,
    "Chia Tohumu": 5,
    "Protein Tozu": 20
}


def menuyu_goster():
    print("\n🍓 YETBAR SMOOTHIE MENÜ 🍌")
    for urun in menu:
        print(f"{urun:20} - {menu[urun]} TL")


def fiyat_hesapla(secimler):
    toplam = 0
    for urun in secimler:
        if urun in menu:
            toplam += menu[urun]
    return toplam


def tarih_al():
    t = str(__import__("time").localtime())
    y, mo, d, h, mi = t[5:9] + t[34:36]
    return f"{y}-{mo}-{d} {h}:{mi}"


def siparisi_kaydet(secimler, toplam_tutar):
    tarih = tarih_al()
    dosya = open("siparis_kayitlari.txt", "a", encoding="utf-8")
    dosya.write(f"Tarih: {tarih}\n")
    dosya.write("Seçilen Ürünler: " + ", ".join(secimler) + "\n")
    dosya.write(f"Toplam Tutar: {toplam_tutar} TL\n")
    dosya.write("-" * 40 + "\n")
    dosya.close()


def satin_al():
    menuyu_goster()
    secimler = []
    print("\nLütfen smoothie'nize eklemek istediğiniz ürünleri yazın.")
    print("İşlemi bitirmek için 'tamam' yazın.\n")

    while True:
        secim = input("Ürün adı: ").strip().title()
        if secim == "Tamam":
            break
        elif secim in menu:
            secimler.append(secim)
            print(f"{secim} eklendi.")
        else:
            print("Menüde böyle bir ürün yok. Tekrar deneyin.")

    if secimler:
        toplam = fiyat_hesapla(secimler)
        print(f"\nSeçilen ürünler: {', '.join(secimler)}")
        print(f"Toplam tutar: {toplam} TL")
        siparisi_kaydet(secimler, toplam)
        print("Siparişiniz kaydedildi. Afiyet olsun!\n")
    else:
        print("Hiç ürün seçilmedi. Sipariş iptal edildi.\n")


def main():
    while True:
        print("\n--- YETBAR OTOMAT ---")
        print("1. Menüyü Görüntüle")
        print("2. Smoothie Satın Al")
        print("3. Çıkış")

        secim = input("Seçiminiz (1-3): ").strip()
        if secim == "1":
            menuyu_goster()
        elif secim == "2":
            satin_al()
        elif secim == "3":
            print("Çıkış yapılıyor. Sağlıklı günler!")
            break
        else:
            print("Hatalı giriş. Lütfen 1, 2 veya 3 girin.")


main()

 
```
