Åimdi beraber kullanÄ±cÄ±lardan aldÄ±ÄŸÄ±mÄ±z verileri iÅŸleyerek, onlara hem Ã§Ä±ktÄ± verecek, hem de aldÄ±ÄŸÄ±mÄ±z ve hesapladÄ±ÄŸÄ±mÄ±z verileri ileride kullanabilmek adÄ±na kaydedeceÄŸiz.


 **Bu bir veri giriÅŸ platformu olacak**


 * kullanÄ±cÄ±ya kaÃ§ giriÅŸ yapacaÄŸÄ±nÄ± soracaÄŸÄ±z
* kullanÄ±cÄ±dan isim-soyisim, yaÅŸ, boy, kilo bilgilerini alarak VKÄ° hesaplayacaÄŸÄ±z
* ardÄ±ndan bunlarÄ± sÄ±nÄ±flara gÃ¶re kaydedeceÄŸiz bir dosyaya kaydedeceÄŸiz


SÄ±nÄ±flar ise ÅŸu ÅŸekilde: 
* 18, 5 kg/m.'nin altÄ±nda olanlar: ZayÄ±f
* 18.5 â€“ 24, 9 kg/m. arasÄ±nda olanlar: Normal kilolu
* 25 â€“ 29, 9 kg/m. arasÄ±nda olanlar: Fazla kilolu
* 30 â€“ 39, 9 kg/m. arasÄ±nda olanlar: Obez
* 40 kg/m.'nin Ã¼zerinde olanlar: Ä°leri derecede obez (morbid obez), olarak gÃ¶rÃ¼lÃ¼r.

```python
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_sinifi(vki):
    if vki < 18.5:
        return "ZayÄ±f"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "Ä°leri derecede obez (morbid obez)"

def veri_kaydet(dosya_adi, veri_listesi):
    with open(dosya_adi, "w", encoding="utf-8") as dosya:
        dosya.write("Ä°sim Soyisim, YaÅŸ, Boy (m), Kilo (kg), VKÄ°, SÄ±nÄ±f\n")
        for veri in veri_listesi:
            dosya.write(f"{veri['isim']} {veri['soyisim']}, {veri['yas']}, {veri['boy']:.2f}, {veri['kilo']}, {veri['vki']:.2f}, {veri['sinif']}\n")

def main():
    veri_listesi = []
    adet = int(input("KaÃ§ kiÅŸi iÃ§in giriÅŸ yapacaksÄ±nÄ±z? "))

    for i in range(adet):
        print(f"\n{i + 1}. kiÅŸi bilgileri:")
        isim = input("Ä°sim: ").strip().capitalize()
        soyisim = input("Soyisim: ").strip().capitalize()
        yas = int(input("YaÅŸ: "))
        boy = float(input("Boy (metre cinsinden, Ã¶rn: 1.75): "))
        kilo = float(input("Kilo (kg cinsinden): "))

        vki = vki_hesapla(kilo, boy)
        sinif = vki_sinifi(vki)

        print(f"{isim} {soyisim} iÃ§in VKÄ°: {vki:.2f} - {sinif}")

        veri_listesi.append({
            "isim": isim,
            "soyisim": soyisim,
            "yas": yas,
            "boy": boy,
            "kilo": kilo,
            "vki": vki,
            "sinif": sinif
        })

    dosya_adi = "vki_kayitlari.txt"
    veri_kaydet(dosya_adi, veri_listesi)
    print(f"\nTÃ¼m veriler '{dosya_adi}' adlÄ± dosyaya kaydedildi.")

if __name__ == "__main__":
    main()
```

```python


menu = {
    "Ã‡ilek": 15,
    "Muz": 10,
    "Yaban Mersini": 18,
    "Ispanak": 8,
    "Badem SÃ¼tÃ¼": 12,
    "Yulaf": 7,
    "Chia Tohumu": 5,
    "Protein Tozu": 20
}


def menuyu_goster():
    print("\nğŸ“ YETBAR SMOOTHIE MENÃœ ğŸŒ")
    for urun in menu:
        print(f"{urun:20} - {menu[urun]} TL")


def fiyat_hesapla(secimler):
    toplam = 0
    for urun in secimler:
        if urun in menu:
            toplam += menu[urun]
    return toplam


def tarih_al():
    t = str(__import__("time").localtime())
    y, mo, d, h, mi = t[5:9] + t[34:36]
    return f"{y}-{mo}-{d} {h}:{mi}"


def siparisi_kaydet(secimler, toplam_tutar):
    tarih = tarih_al()
    dosya = open("siparis_kayitlari.txt", "a", encoding="utf-8")
    dosya.write(f"Tarih: {tarih}\n")
    dosya.write("SeÃ§ilen ÃœrÃ¼nler: " + ", ".join(secimler) + "\n")
    dosya.write(f"Toplam Tutar: {toplam_tutar} TL\n")
    dosya.write("-" * 40 + "\n")
    dosya.close()


def satin_al():
    menuyu_goster()
    secimler = []
    print("\nLÃ¼tfen smoothie'nize eklemek istediÄŸiniz Ã¼rÃ¼nleri yazÄ±n.")
    print("Ä°ÅŸlemi bitirmek iÃ§in 'tamam' yazÄ±n.\n")

    while True:
        secim = input("ÃœrÃ¼n adÄ±: ").strip().title()
        if secim == "Tamam":
            break
        elif secim in menu:
            secimler.append(secim)
            print(f"{secim} eklendi.")
        else:
            print("MenÃ¼de bÃ¶yle bir Ã¼rÃ¼n yok. Tekrar deneyin.")

    if secimler:
        toplam = fiyat_hesapla(secimler)
        print(f"\nSeÃ§ilen Ã¼rÃ¼nler: {', '.join(secimler)}")
        print(f"Toplam tutar: {toplam} TL")
        siparisi_kaydet(secimler, toplam)
        print("SipariÅŸiniz kaydedildi. Afiyet olsun!\n")
    else:
        print("HiÃ§ Ã¼rÃ¼n seÃ§ilmedi. SipariÅŸ iptal edildi.\n")


def main():
    while True:
        print("\n--- YETBAR OTOMAT ---")
        print("1. MenÃ¼yÃ¼ GÃ¶rÃ¼ntÃ¼le")
        print("2. Smoothie SatÄ±n Al")
        print("3. Ã‡Ä±kÄ±ÅŸ")

        secim = input("SeÃ§iminiz (1-3): ").strip()
        if secim == "1":
            menuyu_goster()
        elif secim == "2":
            satin_al()
        elif secim == "3":
            print("Ã‡Ä±kÄ±ÅŸ yapÄ±lÄ±yor. SaÄŸlÄ±klÄ± gÃ¼nler!")
            break
        else:
            print("HatalÄ± giriÅŸ. LÃ¼tfen 1, 2 veya 3 girin.")


main()

 
```
